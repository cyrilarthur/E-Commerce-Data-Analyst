```python
# DATA TRANSFORMATION 
```

```python
# Converting data type to suit time series analysis 
Ecommerce_Dataframe['Date'] = pd.to_datetime(Ecommerce_Dataframe['Date'], format="%d/%m/%Y")
```

```python
# Renaming column(s) for 
Ecommerce_Dataframe.rename({'Date':'ds', 'Transaction_Result':'y'},axis=1,inplace=True)
```

```python
# TRAIN MODEL 
```

```python
m = Prophet(interval_width=0.95,daily_seasonality=True)
model= m.fit(Ecommerce_Dataframe)
```

```python
# FORECASTING 
```

```python
future = m.make_future_dataframe(periods=100,freq='M')
forecast = m.predict(future)
```
